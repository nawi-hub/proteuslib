```python
import matplotlib.pyplot as plt
import numpy as np


data = np.genfromtxt('mem_cost_ele_cost_3stage.csv', skip_header=1, delimiter=',')

num_procs = int(np.sqrt(np.shape(data)[0]))

x_mesh = np.reshape(data[:, 0], (num_procs, num_procs)).T
y_mesh = np.reshape(data[:, 1], (num_procs, num_procs)).T

ec = np.reshape(data[:, 2], (num_procs, num_procs)).T
lcow = np.reshape(data[:, 3], (num_procs, num_procs)).T

plt.figure(figsize=(7, 5), dpi=100)
plt.set_cmap('YlGnBu')
plt.contourf(x_mesh, y_mesh, lcow, np.linspace(1, 2, 11), extend='max')
plt.xlabel('Membrane Cost (\$/m$^2$)')
plt.ylabel('Electricity Cost (\$/kWh)')
plt.yticks(np.linspace(0.03, 0.15, 7))
plt.title('Levelized Cost of Water (\$/m$^3$)')
plt.colorbar()
plt.savefig('mem_cost_elec_cost_3_stage.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('mem_cost_water_perm_3stage.csv', skip_header=1, delimiter=',')

num_procs = int(np.sqrt(np.shape(data)[0]))

x_mesh = np.reshape(data[:, 0], (num_procs, num_procs)).T
y_mesh = np.reshape(data[:, 1], (num_procs, num_procs)).T

ec = np.reshape(data[:, 2], (num_procs, num_procs)).T
lcow = np.reshape(data[:, 3], (num_procs, num_procs)).T

plt.figure(figsize=(7, 5), dpi=100)
plt.set_cmap('YlGnBu')
plt.contourf(x_mesh, y_mesh, lcow, np.linspace(1, 2, 11), extend='max')

plt.xlabel('Membrane Cost (\$/m$^2$)')
plt.ylabel('Water Permeability (1e-12 m/s-Pa)')
plt.yticks(np.linspace(2e-12, 10e-12, 9), np.linspace(2, 10, 9))
plt.title('Levelized Cost of Water (\$/m$^3$)')
plt.colorbar()
plt.savefig('mem_cost_water_perm_3_stage.png')

# plt.plot(data[:, 0], data[:, 1], 'k.', markersize=1)
# plt.savefig('mem_cost_water_perm_3_stage_dots.png')
plt.show()
```
